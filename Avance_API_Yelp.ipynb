#### <span style="color:#00b6ff">**Normalización de datos**</span>

```python
#Se normalizan los datos para poder limpiarlos después

df_plano = pd.json_normalize(data["businesses"])
```

```python
#Se definen nuevas columnas ordenadas
df_final = df_plano[[
    "name",
    "review_count",
    "rating",
    "location.city",
    "location.zip_code",
    "price"
]].copy()

#Cambiamos los nombres para que quede más legible

df_final.columns = ["Nombre", "Reseñas", "Calificación", "Ciudad", "CP", "Precio"]

display(df_final.head())
```

#### <span style="color:#00b6ff">**Limpieza de datos de la API**</span>

```python
print(df_final.isnull().sum())
print(f"\nShape: {df_final.shape}")
print(f"\nTipos de datos:\n{df_final.dtypes}")
```

```python
# Limpieza df_final
moda_precio = df_final['Precio'].mode()[0]
df_final['Precio'] = df_final['Precio'].fillna(moda_precio)

# Verificación
print(f"✅ Nulos en Precio imputados con: '{moda_precio}'")
print(df_final.isnull().sum())
print(f"Shape: {df_final.shape}")
```

#### <span style="color:#00b6ff">**Integración**</span>

#### _Se realiza la integración desde la ciudad, en este caso Miami_

```python
#Utilizamos Merge para unir los resultados del análisis EDA con el filtrado de la API YELP

df_integrado = pd.merge(
    df_clientes, 
    df_final, 
    left_on="ciudad_residencia", 
    right_on="Ciudad", 
    how="left"
)

print("Integración correcta")
print(f"Nuevas columnas añadidas: {df_final.columns.tolist()}")
display(df_integrado.head())
```

#### <span style="color:#00b6ff">**Limpieza de datos**</span>

#### _Valores nulos_

```python
#Cantidad totales de valores nulos 
#Tamaño total de la tabla

print(f"Tamaño total de la tabla: {df_integrado.shape}")
print("Valores nulos por columna:")
print(df_integrado.isnull().sum())
```

```python
df_integrado['rango_etario'] = df_integrado['rango_etario'].fillna('No especifica')
print(f"Tamaño total de la tabla: {df_integrado.shape}")
print("Valores nulos por columna:")
print(df_integrado.isnull().sum())
```

#### <span style="color:#00b6ff">**Filtrado de datos**</span>

#### _A continuación buscaremos los mejores restaurantes en Miami (según la calificación y un número de reseñas) y los mejores clientes (según el gasto promedio y el poder adquisitivo)_ 

```python

df_miami_limpio = df_integrado.copy()

df_top_resto = df_miami_limpio[
    (df_miami_limpio["Calificación"] >= 4.0) & 
    (df_miami_limpio["Reseñas"] > 50)
]

#Buscamos a los clientes que tiene el ingreso por encima del 50%
umbral_ingresos = df_miami_limpio["ingresos_mensuales"].median()
df_objetivo = df_top_resto[df_top_resto["ingresos_mensuales"] >= umbral_ingresos]

print(f"Estos son la cantidad total de combinaciones ideales (cliente-restaurante): {df_objetivo.shape[0]}")
```

#### <span style="color:#00b6ff">**Recomendaciones**</span>

#### _Estos son los 10 restaurantes recomendados para nuestros clientes objetivo_

```python
recomendados = df_objetivo[["Nombre", "Calificación", "Reseñas", "Precio"]].drop_duplicates()

recomendados_ranking = recomendados.sort_values(by=["Calificación", "Reseñas"], ascending=False)

print("Los 10 mejores restaurantes para nuestros clientes en Miami:")
display(recomendados_ranking.head(10))
```
